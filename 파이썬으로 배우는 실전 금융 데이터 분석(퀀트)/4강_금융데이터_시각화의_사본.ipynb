# 2 - seaborn 으로 데이터 분포 살펴보기

- seaborn 임포트,  한글 글꼴 지정

```
# seaborn 임포트,  한글 글꼴 지정
import seaborn as sns
sns.set()
plt.rc('font', family='NanumBarunGothic')
```

```
tips = sns.load_dataset('tips')  # 내장 데이터 셋 불러오기
```

```
tips.head()    # dataframe의 앞부분만 살펴보기
```

```
sns.countplot(data=tips, x='day');
plt.title('요일별 팁을 준 횟수');
```

```
sns.countplot(data=tips,x='time')
```

```
sns.boxplot(data=tips, x='day', y='tip');
plt.title('요일별 팁의 Box Plot');
```

```
sns.boxplot(data=tips,x='smoker',y='tip')
```

```
sns.scatterplot(data=tips, x='total_bill', y='tip');
plt.title('식사 금액과 팁의 관계');
```

```
sns.jointplot(data=tips, x="total_bill", y="tip");
plt.suptitle("식사 금액과 팁의 관계 Joint Plot", y=1.02);
```

```
sns.boxplot(data=tips, x='day', y='tip');
plt.title('요일별 팁의 Box Plot');
```

```
sns.boxplot(data=tips, x='day', y='tip', hue='sex');
plt.title('요일 별, 성별 팁의 Box Plot');
```

```
sns.violinplot(data=tips, x="day", y="tip", hue="sex")
plt.title("요일 별, 성별 팁의 Violin Plot")
plt.show()
```

```
sns.distplot(tips.tip, kde=True, rug=True);
plt.title('팁의 분포');
```

### <연습문제>
성별 별로, 팁을 준 횟수를 차트로 그려보세요

```
ax=sns.countplot(data=tips,x='sex')
plt.title('성별 팁을 준 횟수')
# countplot에 값 표시
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 10, height, ha = 'center', size = 12)
ax.set_ylim(0, 180)
```

### Line Plot 그리기

```
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))
ts = ts.cumsum()
ts.plot(figsize=(12,8));
```

```
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list("ABCD"))
df = df.cumsum()
df.plot();
```

```
adj_close = pd.read_csv("stock.adj_close.csv", index_col=0, encoding='cp949', parse_dates=True)
```

```
itemname = pd.read_csv("stockinfo.itemname.csv", index_col=0, encoding='cp949')
sector = pd.read_csv("stockinfo.sector.csv", index_col=0, encoding='cp949')

stock_info = pd.concat([itemname, sector], axis=1)
```

```
# 데이터 준비
df_a = stock_info
df_b = adj_close
```

```
df_a
```

```
df_b
```

```
my_concat_df = pd.concat([df_a, df_b], axis=1)
my_concat_df
```

```
# 섹터의 빈도수를 구합니다.
sector_count = my_concat_df["Sector"].value_counts()
sector_count
```

```
sector_count.plot(kind='barh', figsize=(10, 8), title="Sector 빈도수")
```

```
# Seaborn countplot 활용하여 빈도수를 표현합니다.
plt.figure(figsize=(10,8))
sns.countplot(data=my_concat_df[my_concat_df["Sector"].isin(sector_count.index)], 
              y="Sector", order=sector_count.index).set_title('섹터별 종목수')
```

```
sector_count.plot(kind='pie', figsize=(10, 8), title="Sector 빈도수")
```
